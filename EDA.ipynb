```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_csv('data/cleaned_data.csv')
df.head(5)
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8, 5))
sns.histplot(df['Average_Player_Age'], kde=True, bins=20)
plt.title('Distribution of Average Player Age')
plt.xlabel('Average Player Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Country', y='Total_Goals_Last_Season', data=df)
plt.title('Total Goals Last Season by Country')
plt.xlabel('Country')
plt.ylabel('Total Goals Last Season')
plt.show()
```

```python
# Distribution of Stadium Capacity
plt.figure(figsize=(10, 6))
sns.histplot(df['Stadium_Capacity'], kde=True, bins=20)
plt.title('Distribution of Stadium Capacity')
plt.xlabel('Stadium Capacity')
plt.ylabel('Frequency')
plt.show()
```

Detailed Analysis

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Team_Formation', data=df, order=df['Team_Formation'].value_counts().index)
plt.title('Distribution of Team Formations')
plt.xlabel('Team Formation')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Stadium_Capacity'], kde=True, bins=20)
plt.title('Distribution of Stadium Capacity')
plt.xlabel('Stadium Capacity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Playing_Style', data=df, order=df['Playing_Style'].value_counts().index)
plt.title('Distribution of Playing Styles')
plt.xlabel('Playing Style')
plt.ylabel('Count')
plt.show()
```

```python
numerical_df = df.select_dtypes(include=[np.number])
corr = numerical_df.corr()
plt.figure(figsize=(16, 12))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
